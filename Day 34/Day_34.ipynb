#  **Working With Dates**

```python
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

## 1.Extract years

```python
date['date_year']=date['date'].dt.year

date.head()
```

## 2.Extract months

```python
date['date_month']=date['date'].dt.month

date.head()
```

## 3.Extract month name




```python
date['date_month_name']=date['date'].dt.month_name()

date.head()
```

## 4.Extract days

```python
date['date_day']=date['date'].dt.day
date.head()
```

## 4.Extract days name

```python
date['date_day_name']=date['date'].dt.day_name()
date.head()
```

## 4.Extract days of week

```python
date['date_day_of_week']=date['date'].dt.day_of_week


date.drop(columns=['product_id','city_id','orders'])
date.head()
```

```python
# is Weekend

date['date_is_weekend']=np.where(date['date_day_name'].isin(['Sunday','Saturday']),1,0)

date.head()
```

## Extract Week

```python
date['date_week']=date['date'].dt.isocalendar().week
date.head()
```

## 6.Extract Quarter

```python
date['date_quarter']=date['date'].dt.quarter
date.head()
```

## 7.Extract semester

```python
date['date_semester']=np.where(date['date_quarter'].isin([1,2]),1,2)
date.head()
```

## Extract time between two dates

```python
import datetime

today=datetime.datetime.today()

today
```

```python
today-date['date']
```

```python
(today-date['date']).dt.days
```

```python
#month passed

((today - date['date']).dt.days / 30).round()
```
