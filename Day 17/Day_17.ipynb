```python
import requests
import pandas as pd

all_data = []

url = "https://weatherbit-v1-mashape.p.rapidapi.com/forecast/3hourly"

headers = {
    "x-rapidapi-host": "weatherbit-v1-mashape.p.rapidapi.com",
    "x-rapidapi-key": "2911aab065msh1001b0054442f7ap1e0740jsnaec496ef19f4"
}

# loop through pages
for page in range(1, 500):   # page 1 to 3
    querystring = {
        "lat": "35.5",
        "lon": "-78.5",
        "units": "imperial",
        "lang": "en",
        "page": page
    }

    response = requests.get(url, headers=headers, params=querystring)

    data = response.json()

    # some APIs may not have 'data', so we check
    if "data" in data:
        all_data.extend(data["data"])

# convert to dataframe
df = pd.DataFrame(all_data)

print(df.head())
```

```python
df.to_csv('weather1.csv')
```
