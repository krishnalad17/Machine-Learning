```
import pandas as pd
import numpy as np
```

```
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
```

```
df = pd.read_csv('/content/covid_toy.csv')
```

```
df.head()
```

```
df.isnull().sum()
```

```
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2,random_state=0)
```

# **1. Aam Zindagi**

```
si = SimpleImputer()

X_train_fever=si.fit_transform(X_train[['fever']])
X_test_fever=si.fit_transform(X_test[['fever']])

X_train_fever.shape #as fever class had 10 missing values
```

```
oe  = OrdinalEncoder(categories=[['Mild','Strong']],dtype=np.int32)

X_train_cough=oe.fit_transform(X_train[['cough']])
X_test_cough=oe.fit_transform(X_test[['cough']])

X_train_cough.shape
```

```
ohe = OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32)

X_train_gender_city=ohe.fit_transform(X_train[['gender','city']])
X_test_gender_city=ohe.fit_transform(X_test[['gender','city']])

X_train_gender_city.shape
```

```
X_train_age = X_train.drop(columns=['gender','fever','cough','city']).values
X_test_age = X_test.drop(columns=['gender','fever','cough','city']).values

X_train_age.shape
```

```
X_train_transformed=np.concatenate((X_train_age,X_train_fever,X_train_gender_city,X_train_cough),axis=1)
X_test_transformed=np.concatenate((X_test_age,X_test_fever,X_test_gender_city,X_test_cough),axis=1)

X_train_transformed.shape
```

# **2. Mentos Zindagi**

```
from sklearn.compose import ColumnTransformer
```

```
col = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']],dtype=np.int32),['cough']),
    ('tnf3',OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32),['gender','city'])
],remainder='passthrough')
```

```
X_train = col.fit_transform(X_train)
X_train.shape
```
