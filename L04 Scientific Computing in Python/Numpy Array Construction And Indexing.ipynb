creating identity matrix

```python
np.eye(3,3)
```

indentity matrix with values

```python
np.diag((4,4,4))  #similar to np.eye just argumnets are values in diagonal
```

numpy.arange() function creates an array of evenly spaced values within a given interval

```python
np.arange(0,100,2)  # 0 is start ,100 is stop ,2 is steps similar to range data
```

linspace() function in NumPy returns an array of evenly spaced numbers over a specified range

```python
np.linspace(1,20,num=100)
```

## Indexing Arrays

```python
arry=np.array([1,2,3])
arry[0]
```

### slicing

```python
arry[:2]
```

### 2D Array Indexing

```python
arr=np.array([[1,2,3],
             [4,5,6]])

arr[0,0] #upper right
```

```python
arr[-1,-1] #lower right
```

```python
arr[0,:] #first row 
```

```python
arr[:,0] #first column
```

```python
arr[:,:2] #first two column
```
