## Random Number Generators

```python
import numpy as np
np.random.rand(10) # returns n random numbers
```

```python
np.random.seed(123) # fixes the random number as it was changing again and again
np.random.rand(10)
```

```python
from sklearn.neighbors import KNeighborsClassifier

np.random.seed(123)
y=np.array([0]*50 + [1]*50)
X=np.random.rand(100,2)

knn_model=KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X,y)
knn_model.score(X,y)
```

```python
rng1=np.random.RandomState(seed=123) # similar to random.rand()
rng1.rand(3)
```

```python
rng1=np.random.RandomState(seed=123)
np.random.rand(2) # this is not affected by RandomState and will generate random number every time
```

```python
rng2=np.random.default_rng(seed=123) # similar to all upper options
rng2.random(3)
```
