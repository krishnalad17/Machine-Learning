## NumPy Advanced Indexing -- Memory Views and Copies

```python
import numpy as np

ary=np.array([[1,2,3],[4,5,6]])
first_row=ary[0]
first_row
```

```python
first_row+=99 #modified values of arrays also with the first_rows
ary
```

```python
ary1=np.array([[1,2,3],[4,5,6]])

second_row=ary1[1].copy() #this hepls to keep the orginal array as it is
second_row+=99
ary1
```

Slicing helps in creating copy 

```python
ary1[:,[0,2]] # this type of slicing is not possible with lists in python  also creates copy

this_is_copy=ary1[:,[0,2]]
this_is_copy+=99
ary1
this_is_copy
```

```python
greater_mask=ary1 >3 # gives boolean values for the value greater than the given number
greater_mask
```

```python
ary1[greater_mask] #tkes the indexing as argument and return the value for True
```

```python
ary1[(ary1 >3 ) & (ary1%2==0)]
```
