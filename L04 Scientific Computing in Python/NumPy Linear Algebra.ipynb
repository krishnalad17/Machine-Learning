## Linear Algebra with Numpy Arrays

```python
import numpy as np
```

Row vector has 1 row and multiple columns

```python
row_vector=np.array([1,2,3]) 
row_vector.shape
```

```python
column_vector=np.array([1,2,3])
column_vector
```

```python
column_vector[:,np.newaxis]
```

both above and below are rehaping techniques

```python
column_vector.reshape(-1,1)
```

```python
row_vector[np.newaxis,:,np.newaxis,np.newaxis].shape
```

None behaves like np.newaxis

```python
row_vector[:,None]
```

matrix multiplication

```python
matrix=np.array([[1,2,3],
                [4,5,6]])

column_vector=np.array([[1,2,3]])
column_vector.reshape(-1,1)
column_vector
```

```python
np.matmul(matrix,row_vector)
```

```python
np.dot(row_vector,row_vector)
```

```python
np.dot(matrix,row_vector)
```

Transpose

```python
matrix.transpose()
```

```python
np.dot(matrix,matrix.transpose())
```
