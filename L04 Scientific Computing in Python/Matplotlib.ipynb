## Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x))
plt.show()
```

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x))

plt.xlim([2,8])
plt.ylim([0,0.75])

plt.xlabel('x-axis')
plt.ylabel('y-axis'); # ; = plt.show()
```

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),label=('sin(x)'))
plt.plot(x,np.cos(x),label=('cos(x)'),color='red')

plt.xlabel('x')
plt.ylabel('f(x)')

plt.legend(loc='upper left')
plt.show()
```

## Scatter Plots

```python
rng=np.random.RandomState(123)
x=rng.normal(size=500)
y=rng.normal(size=500)

plt.scatter(x,y)
plt.show()
```

## Bar Plot

```python
means=[5,8,10]
std=[0.2,0.4,0.5]
bar_labels=['bar1','bar2','bar3']

xpos=list(range(len(bar_labels)))
plt.figure(figsize=(3,2))
plt.bar(xpos,means,yerr=std)
plt.show()
```

## Histograms

```python
rng=np.random.RandomState(123)
x=rng.normal(0,20,1000)

bins=np.arange(-100,100,5)

plt.hist(x,bins=bins)
plt.show()
```

```python
rng=np.random.RandomState(123)
x1=rng.normal(0,20,1000)
x2=rng.normal(15,10,1000)

bins=np.arange(-100,100,5)

plt.hist(x1,bins=bins,alpha=0.5)
plt.hist(x2,bins=bins,alpha=0.5)
plt.show()
```

## Subplots

```python
x=range(11)
y=range(11)

fig,ax=plt.subplots(nrows=2,ncols=3,sharex=True,sharey=True)

for row in ax:
    for col in row:
        col.plot(x,y)

plt.show()
```

## Colour and Marker

```python
x=np.linspace(0,10,100)

plt.plot(x,np.sin(x),color='grey',marker='^',linestyle='')
plt.savefig('krishnaplot.pdf')
plt.show()
```
