```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('/content/train.csv')[['Age','Pclass','SibSp','Parch','Survived']]
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
X=df.iloc[:,0:4]
y=df['Survived']
```

```python
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,cv=20,scoring='accuracy'))
```

# **Applying Feature Construction**

```python
X['Family']=X['Parch'] + X['SibSp'] + 1
```

```python
X.head()
```

```python
def myfunc(num):
  if num==1:#alone
    return 0
  elif num >1 and num <=4:#small family
    return 1
  else:#large family
    return 2
```

```python
myfunc(5)
```

```python
X['Family_type']=X['Family'].apply(myfunc)
```

```python
X.head()
```

```python
X.drop(columns=['SibSp','Parch','Family'],inplace=True)
```

```python
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,cv=20,scoring='accuracy'))
```
