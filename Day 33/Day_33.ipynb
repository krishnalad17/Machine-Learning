```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head(4)
```

```python
df['number'].unique()
```

```python
fig=df['number'].value_counts().plot.bar()
fig.set_title("Passengers Travelling With")
```

```python
df['number_numerical']=pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
df['number_cat']=np.where(df['number_numerical'].isnull(),df['number'],np.nan)
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['cab_num']=df['Cabin'].str.extract(r'(\d+)') #captures umeric part
df['cab_cat']=df['Cabin'].str[0] #captures first letter
```

```python
df.head()
```

```python
df['cab_cat'].value_counts().plot(kind='bar')
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(20)
```
